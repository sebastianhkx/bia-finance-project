```python
import numpy as np
import pandas as pd
import datetime
import matplotlib.pyplot as plt 
%matplotlib inline
import matplotlib
matplotlib.rc('figure', figsize=(8, 7))
matplotlib.__version__
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
```

```python
amazon = pd.read_csv("AMZN.csv", parse_dates=True, index_col='Date')               
amazon.head()
```

```python
apple = pd.read_csv("AAPL.csv",parse_dates=True, index_col='Date')
apple.head()
```

```python
google = pd.read_csv("GOOGL.csv", parse_dates=True, index_col='Date')
google.head()
```

```python
bp = pd.read_csv('BP.csv', parse_dates=True, index_col='Date')
bp.head()
```

```python
costco = pd.read_csv("COST.csv", parse_dates=True, index_col='Date')
costco.head()
```

```python
sp500 = pd.read_csv("^GSPC.csv", parse_dates=True, index_col='Date')
sp500.head()
```

```python
# EDA
stocks = pd.DataFrame({"^GSPC" : sp500["Adj Close"],
                       "APPLE": apple["Adj Close"],
                      "AMAZON": amazon["Adj Close"],
                      "GOOGLE": google["Adj Close"],
                      "BP": bp["Adj Close"],
                      "COSTCO" : costco["Adj Close"]})

stocks.head()
```

```python
stocks.plot(grid=True)
```

```python
stock_return = stocks.apply(lambda x: x / x[0])
stock_return.head() - 1
```

```python
stock_return.plot(grid = True).axhline(y = 1, color = "black", lw = 2)
```

```python
# finding the beta of the stocks using regression
prices_amzn = pd.concat([amazon['Adj Close'], sp500['Adj Close']], axis=1)
prices_amzn.columns = ['AMZN', '^GSPC']

# check the head of the dataframe
print(prices_amzn.head())

# calculate monthly returns
returns_amzn = prices_amzn.pct_change(1)
clean_returns_amzn = returns_amzn.dropna(axis=0)  # drop first missing row
print(clean_returns_amzn.head())
```

```python
# split dependent and independent variable
X = clean_returns_amzn ['AMZN'] - clean_returns_amzn ['^GSPC']
y = clean_returns_amzn ['AMZN'] 

# Add a constant to the independent value
X1 = sm.add_constant(X)

# make regression model 
model = sm.OLS(y, X1)

# fit model and print results
results = model.fit()
print(results.summary())
```
